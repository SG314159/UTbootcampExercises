```python
import pandas as pd
```

```python
# File paths for files to load
school_to_load = "Resources/schools_complete.csv"
student_to_load = "Resources/clean_students_complete.csv"
```

```python
# Read in csv data into dataframes
school_data_df = pd.read_csv(school_to_load)
school_data_df
```

```python
student_data_df = pd.read_csv(student_to_load)
student_data_df.head(10)
```

```python
# Starting here with Section 4.7.1 Merging
data_complete = pd.merge(student_data_df,school_data_df,on=["school_name","school_name"])
data_complete.head(10)
```

```python
 # Get total number of students; counts non-null values for each column
student_count = data_complete["Student ID"].count()
student_count
```

```python
# Calculate the total number of schools 
school_count = data_complete["school_name"].nunique()
# Also: len(data_complete["school_name"].unique())
school_count
```

```python
# Calculate total budget for all schools
# Note to use school_data_df and not data_complete df
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
# Calculate the mean reading and math scores
avg_reading = data_complete["reading_score"].mean()
avg_reading
```

```python
avg_math = data_complete["math_score"].mean()
avg_math
```

```python
# Create dataframes students who passed (>=70) for reading and math
passed_math = data_complete[ data_complete["math_score"]>=70 ]
passed_math.head()
```

```python
passed_reading = data_complete[ data_complete["reading_score"]>=70 ]
passed_reading.head()
```

```python
# One of many ways to get number of students who passed
passed_math_count = passed_math["student_name"].count()
passed_math_count
```

```python
passed_reading_count = passed_reading["student_name"].count()
passed_reading_count
```

```python
# Calculate % passing each test
passed_math_percentage = passed_math_count/student_count * 100
passed_math_percentage
```

```python
passed_read_percentage = passed_reading_count/student_count * 100
passed_read_percentage
```

```python
# Get number of students who passed both math and reading
passed_both = data_complete[ (data_complete["reading_score"]>=70) & (data_complete["math_score"]>=70) ]
passed_both
```

```python
passed_both_count = passed_both["student_name"].count()
passed_both_count
```

```python
# Calculate % passing both tests
passed_both_percentage = passed_both_count/student_count * 100
passed_both_percentage
```

```python
# Section 4.7.7 - Creating new dataframe for district summary
# Adding a list of values with keys to create a new DataFrame.
district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": avg_math,
          "Average Reading Score": avg_reading,
          "% Passing Math": passed_math_percentage,
         "% Passing Reading": passed_read_percentage,
        "% Overall Passing": passed_both_percentage}])
district_summary_df
```

```python
# Section 4.7.8
# map function intro
my_grades = ['B','C','B','D']
my_grades = pd.Series(my_grades)
my_grades
```

```python
my_grades = my_grades.map({'B':85,'C':77,'A':95})
my_grades
```

```python
# Using the format() function. Will format each grade to nearest whole number.
my_grades = [92.25, 84.45, 86.78, 98.25]

for grade in my_grades:
    print("{:.1f}".format(grade))
```

```python
# Convert the numerical grades to a Series.
my_grades = pd.Series([92.34, 84.56, 86.78, 78.32])
my_grades
```

```python
# Format the grades to the nearest whole number percent.
my_grades = my_grades.map("{:.0f}".format)
my_grades
```
